```python
import pandas as pd
import nltk 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading the data
messages = pd.read_csv('./data/SMSSpamCollection.txt', sep = '\t', names = ['label', 'message'])
```

```python
messages.head()
```

```python
messages.groupby('label').describe()
```

```python
messages['length'] = messages['message'].apply(len)
```

```python
messages.head()
```

```python
messages.info()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="label", data=messages)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14,4))
sns.histplot(messages.length.values, bins=70, kde=True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(messages.length.values, range(messages.shape[0]))
plt.xlabel('length', fontsize=12)
plt.ylabel('count', fontsize=12)
plt.show()
```
